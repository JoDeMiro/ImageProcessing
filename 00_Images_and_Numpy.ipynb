# 00 - Képek mint NumPy tömbök

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io
%matplotlib inline
```

```python
pic = skimage.io.imread('https://github.com/JoDeMiro/ImageProcessing/blob/main/images/woman2.jpg?raw=true')

print(type(pic))

print(pic.shape)

red   = pic[:, :, 0]
green = pic[:, :, 1]
blue  = pic[:, :, 2]
```

```python

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 8))

ax1.imshow(pic, cmap='gray'); ax1.set_title('RGB')
ax2.imshow(red, cmap='gray'); ax2.set_title('Red')
ax3.imshow(green, cmap='gray'); ax2.set_title('Green')
ax4.imshow(blue, cmap='gray'); ax2.set_title('Blue')

plt.show()
```
