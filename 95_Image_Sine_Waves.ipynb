```python
import numpy as np
import matplotlib.pyplot as plt

import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual

import warnings
warnings.filterwarnings('ignore')
```

```python
def demo(A, k, theta, linspace = 360):

  x = np.arange(0, linspace, 0.1)

  y = A * np.sin(k*x/180*np.pi + theta)

  img = np.tile(y, (150, 1))

  fig, (a1, a2) = plt.subplots(2, 1, figsize=(12,6))
  a1.imshow(img)
  a2.plot(x, y)
  plt.tight_layout()
  plt.show()
  
  return (A, k, theta, linspace)
```

```python
#@title
a = widgets.FloatSlider(min=0, max=30, step=0.1, value=1)
b = widgets.IntSlider(min=1, max=360, step=1, value=1)
c = widgets.FloatSlider(min=0, max=6.28, step=0.01, value=0)
d = widgets.IntSlider(min=100, max=360, step=1, value=360)
ui = widgets.HBox([a, b, c, d])

out = widgets.interactive_output(demo, {'A': a, 'k': b, 'theta': c, 'linspace': d})

display(ui, out)
```

```python
def demotwo(A, k, theta, linspace = 360):

  x = np.arange(0, linspace, 0.1)

  y1 = A * np.sin(k*x/180*np.pi + theta)
  y2 = A/3 * np.sin(3*k*x/180*np.pi + theta)
  y  = (y1 + y2)/2

  img = np.tile(y, (150, 1))

  fig, (a1, a2) = plt.subplots(2, 1, figsize=(12,6))
  a1.imshow(img)
  a2.plot(x, y1)
  a2.plot(x, y2)
  a2.plot(x, y, label='composit')
  plt.tight_layout()
  plt.legend()
  plt.show()
  
  return (A, k, theta, linspace)
```

```python
#@title
a = widgets.FloatSlider(min=0, max=30, step=0.1, value=1)
b = widgets.IntSlider(min=1, max=20, step=1, value=10)
c = widgets.FloatSlider(min=0, max=6.28, step=0.01, value=0)
d = widgets.IntSlider(min=100, max=360, step=1, value=360)
ui = widgets.HBox([a, b, c, d])

out = widgets.interactive_output(demotwo, {'A': a, 'k': b, 'theta': c, 'linspace': d})

display(ui, out)
```

```python
#@title
def demothree(A, k, theta, linspace = 360, alpha = 100):

  alpha = alpha / 100

  x = np.arange(0, linspace, 0.1)

  y1 = A * np.sin(k*x/180*np.pi + theta)
  y2 = A/3 * np.sin(3*k*x/180*np.pi + theta)
  y3 = A/5 * np.sin(5*k*x/180*np.pi + theta)
  y  = (y1 + y2 + y3)/3

  img = np.tile(y, (150, 1))

  fig, (a1, a2) = plt.subplots(2, 1, figsize=(12,6))
  a1.imshow(img)
  a2.plot(x, y1, alpha=alpha)
  a2.plot(x, y2, alpha=alpha)
  a2.plot(x, y3, alpha=alpha)
  a2.plot(x, y, label='composit')
  plt.tight_layout()
  plt.legend()
  plt.show()
  
  return (A, k, theta, linspace)
```

```python
#@title
a = widgets.FloatSlider(min=0, max=30, step=0.1, value=1)
b = widgets.IntSlider(min=1, max=20, step=1, value=4)
c = widgets.FloatSlider(min=0, max=6.28, step=0.01, value=0)
d = widgets.IntSlider(min=100, max=360, step=1, value=360)
e = widgets.IntSlider(min=0, max=100, step=1, value=30)
ui = widgets.HBox([a, b, c, d, e])

out = widgets.interactive_output(demothree, {'A': a, 'k': b, 'theta': c, 'linspace': d, 'alpha': e})

display(ui, out)
```
